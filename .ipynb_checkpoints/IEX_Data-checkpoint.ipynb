```python
jpm = pd.read_csv('./data/jpm.csv', index_col=[0])
jpm.fillna(method='ffill', inplace=True)
```

```python
jpm_low = go.Scattergl(
    x = jpm.index.tolist(),
    y = jpm['low'].tolist(),
    name = 'JPM_LOW'
)

jpm_average = go.Scattergl(
    x = jpm.index.tolist(),
    y = jpm['average'].tolist(),
    name = 'JPM_AVG'
)

jpm_high = go.Scattergl(
    x = jpm.index.tolist(),
    y = jpm['high'].tolist(),
    name = 'JPM_HIGH'
)

layout = go.Layout(
    title = 'JPM Stock Price'
)

fig = go.Figure([jpm_low, jpm_average, jpm_high], layout)

iplot(fig)
```
