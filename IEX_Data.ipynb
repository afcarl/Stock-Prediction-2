```python
amzn = amazon_daily[['date', 'close']]
```

```python
amzn.set_index('date', inplace=True)
```

```python
amzn.shift(1)
```

```python
lags = 10
for i in range(1, lags+1):
    tmp_close = amzn.shift(i)['close']
    tmp_close_col = f'close_{i}'
    amzn[tmp_close_col] = tmp_close
```

```python
amzn.dropna(inplace=True)
```

```python
amzn.head()
```

# XGBoost

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
model = XGBRegressor()
```

```python
X = amzn.drop('close', axis=1)
y = amzn['close']
```

```python
def train_test(X, y, test_size):
    X_train = X[test_size:]
    X_test = X[-test_size:]
    
    y_train = y[test_size:]
    y_test = y[-test_size:]
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = train_test(X, y, test_size=30)
```

```python
model.fit(X_train, y_train)
```

```python
y_preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_preds))
mae = mean_absolute_error(y_true=y_test, y_pred=y_preds)
print(f'RMSE: {rmse}\nMAE: {mae}')
```

```python
x = X_test.index

amzn_actuals = go.Scattergl(
    x = x,
    y = y_test.values,
    name = 'Actuals'
)

amzn_forecast = go.Scattergl(
    x = x,
    y = y_preds,
    name = 'Forecast'
)

layout = go.Layout(
    title = 'Amazon Stock Price [Daily] Forecasted'
)

fig = go.Figure([amzn_actuals, amzn_forecast], layout)

iplot(fig)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
model = Sequential()
model.add(Dense(300))
model.add(Dropout(0.5))
model.add(Dense(1))
model.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.MSE)
```

```python
history = model.fit(X_train.values, y_train.values, epochs = 50, 
                    batch_size=30, validation_data=(X_test.values, y_test.values))
```

```python
y_preds = model.predict(X_test.values)
```

```python
y_preds = [y_preds[i][0] for i in range(len(y_preds))]
```

```python
rmse = np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_preds))
mae = mean_absolute_error(y_true=y_test, y_pred=y_preds)
print(f'RMSE: {rmse}\nMAE: {mae}')
```
